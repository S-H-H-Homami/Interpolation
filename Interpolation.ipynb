# Interpolation 1D



Libraries

```python
from scipy import interpolate
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import pylab as py
```

### Creating our datasets

```python
#Data01
data01=pd.read_table('/content/data01.txt', header=None, delimiter=" ", names= ['X', 'Y'])
df01=pd.DataFrame(data01)
X01= df01['X']
y01= df01['Y']
x_val01=[x+0.5 for x in X01]
# x_val01=[x+0.5 for x in X01[:len(X01)-1]] can be used in case extrapolation is not needed.

#Data02
data02=pd.read_table('/content/data02.txt', header=None, delimiter=" ", names= ['X', 'Y'])
df02=pd.DataFrame(data02)
X02= df02['X']
y02= df02['Y']
x_val02=[x+0.5 for x in X02]


#Data03
data03=pd.read_table('/content/data03.txt', header=None, delimiter=" ", names= ['X', 'Y'])
df03=pd.DataFrame(data03)
X03= df03['X']
y03= df03['Y']
x_val03=[x+0.5 for x in X03]


#Data04
data04=pd.read_table('/content/data04.txt', header=None, delimiter=" ", names= ['X', 'Y'])
df04=pd.DataFrame(data04)
X04= df04['X']
y04= df04['Y']
x_val04=[x+0.05 for x in X04]
```

## 1d interpolation with interp1d and Rbf

```python
func01=interpolate.interp1d(X01, y01, fill_value= 'extrapolate')
func_rbf01=interpolate.Rbf(X01, y01, fill_value= 'extrapolate')
plt.plot(X01, y01, label='Real Data')
plt.plot(x_val01, func_rbf01(x_val01), label='RBF')
plt.plot(x_val01, func01(x_val01), label='Interp1d')
plt.scatter(X01, y01, c='black')
plt.title('interpolate')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
```

```python
func02=interpolate.interp1d(X02, y02, fill_value= 'extrapolate')
plt.plot(X02, y02, label='Real Data')
plt.plot(x_val02, func02(x_val02), label='Interp1d')
plt.scatter(X02, y02, c='black')
plt.title('interpolate')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
```

# Interpolation 2D


Generating New Data

```python
X2d= np.arange(-5, 5)
y2d= np.arange(-5, 5)

x_val_2d=np.arange(-4.5, 4.5, 0.1)
y_val_2d=np.arange(-4.5, 4.5, 0.1)
```

Creating another dimension

```python
def Z(x, y):
  return np.power(x, 2)+np.power(y, 2)
```

Interpolation

```python
func_2d_01=interpolate.interp2d(X2d, y2d, Z(X2d, y2d))

py.imshow(func_2d_01(x_val_2d, y_val_2d), cmap= py.cm.jet)
```
